# FIF Test

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from scipy.optimize import least_squares

import FIF_test as ft
```

Create a FIF of three maps

```python
A_FIF_3, b_FIF_3 = ft.inter_gen(np.array([0,0.3,0.7,1]),np.array([0,0.2,0.5,0]),np.array([0.8,0.3,-0.25]),inc_a = False)
```

```python
FIF_3 = ft.chaos(A_FIF_3,b_FIF_3)
FIF_3 = FIF_3[FIF_3[:,0].argsort()]
plt.figure(figsize = (20,10))
plt.plot(FIF_3[:,0],FIF_3[:,1]) 
```

Find the moments of this FIF, and a nearby starting point, This starting point was chosen with the correct parameters, except the x co-ordiniate interpolation points were evenly spread apart. We hope that the moment method is flexible enough to deduce the location of these points.

```python
M_FIF_3 = ft.mom_wrap(A_FIF_3,b_FIF_3,moment_calc = True)

print(M_FIF_3)

x_ex = np.array([0.3,0.7,0.2,0.5,0,0.8,0.3,-0.25])

x0 = np.array([0.33,0.66,0.2,0.5,0,0.8,0.3,-0.25])

print(x0)
```

Set-up an appropriate cost function 

```python
k1 = 0

def FIF_3_cost(x):
    
    global k1
    
    if k1%10 == 0 :
        print('Current map parameters:')
        print('x points:')
        print(np.hstack( ([0],x[:2],[1])) )
        print('y points:')
        print(np.hstack( ([0],x[2:5])))
        print('d values:')
        print(x[5:])


    k1+=1
    
    return np.hstack( ( (ft.FIF_wrap(x)@M_FIF_3 - M_FIF_3)))
```

```python
FIF_3_cost(x0)
```

Take result of the optimisation

```python
print('Approximant:')
print('x points:')
x_ap = np.hstack( ([0],FIF_ls[:2],[1]))
print(x_ap )
print('y points:')
y_ap = np.hstack( ([0],FIF_ls[2:5]))
print(y_ap)
print('d values:')
d_ap = FIF_ls[5:]
print(d_ap)

A_app_3, b_app_3 = ft.inter_gen(x_ap,y_ap,d_ap,inc_a = False)

FIF_3_ap = ft.chaos(A_app_3,b_app_3)
FIF_3_ap = FIF_3_ap[FIF_3_ap[:,0].argsort()]
```

Now we look for only a local minima with more randomness added. 


This is the fun part, be warned it is addictive to repeatly run this with different starting points and see what comes out. Here we select a perturbation of the real answer and stop the optimisation early so not even a minima is found. This forms experimental evidence of the Collage Theorem for moments.

```python
k1 = 0

x_new0 = x0 + 0.1*np.random.randn(len(x0)) 
x_new0[4] = 0

bnds = ([0.1,0.5,-np.inf,-np.inf,-0.01,-1,-1,-1], [0.5,0.9,np.inf,np.inf,0.01,1,1,1])

sol_FIF_3 = least_squares(FIF_3_cost,x_new0 ,bounds= bnds,verbose = 2,max_nfev = 30)

print(sol_FIF_3.x)
```

```python
print('Approximant:')
print('x points:')
x_ap = np.hstack( ([0],sol_FIF_3.x[:2],[1]))
x_in = np.hstack( ([0],x_new0[:2],[1]))
print(x_ap )
print('y points:')
y_ap = np.hstack( ([0],sol_FIF_3.x[2:5]))
y_in = np.hstack( ([0],x_new0[2:5]))
print(y_ap)
print('d values:')
d_ap = sol_FIF_3.x[5:]
d_in = x_new0[5:]
print(d_ap)

A_app_3, b_app_3 = ft.inter_gen(x_ap,y_ap,d_ap,inc_a = False)
A_in_3, b_in_3 = ft.inter_gen(x_in,y_in,d_in,inc_a = False)


FIF_3_ap = ft.chaos(A_app_3,b_app_3)
FIF_3_ap = FIF_3_ap[FIF_3_ap[:,0].argsort()]

FIF_3_in = ft.chaos(A_in_3,b_in_3)
FIF_3_in = FIF_3_in[FIF_3_in[:,0].argsort()]
```

```python
from scipy.spatial.distance import directed_hausdorff
def Haus_dist(U,V):
    
    return max(directed_hausdorff(U,V)[0],directed_hausdorff(V,U)[0])

print('Hausdorff distance: Initial starting point')
print(Haus_dist(FIF_3,FIF_3_in))
print('Hausdorff distance: Approximant')
print(Haus_dist(FIF_3,FIF_3_ap))
plt.figure(figsize = (20,10))
plt.plot(FIF_3[:,0],FIF_3[:,1], label = 'Function') 
plt.plot(FIF_3_ap[:,0],FIF_3_ap[:,1], label = 'Approximant') 
plt.plot(FIF_3_in[:,0],FIF_3_in[:,1], label = 'Initial Guess') 
plt.legend(fontsize = 20)
```
