```python
import numpy as np
from scipy.optimize import least_squares
import matplotlib.pyplot as plt
from PIL import Image
from scipy import optimize

#import extras for our examples
import Steem_ex as se
import Moment_Function as mf

%matplotlib inline
```

Set-up Steemson triangle example by forming the IFS for a Steemson triangle with side lengths $1,1.1,0.85$.
The actual functions used are in Python files, we simply import them here.

```python
#start Steemson attempt
a  =  1.1
b  =  0.85

imgxy = 500

#co ordinates of triangles third vertex
Cy  =  (0.5*np.sqrt(-1 + 2*a**2 - a**4 + 2*b**2 + 2*a**2*b**2 - b**4))
Cx  =  (0.5*(1 - a**2 + b**2))

#angles for roatation
cosA  =  (Cx/b)
sinA  =  (Cy/b)
cosB  =  ((1.0-Cx)/a)
sinB  =  (Cy/a)

#scaling factors 
alphaNFF  =  (b)/((a**2 + b**2))
betaNFF   =  (a)/(a**2 + b**2)
gammaNFF  =  -1.0*(1.0-a**2-b**2)/(a**2 + b**2)

alphaNNF  =  (b/(1.0 + b**2))
betaNNF   =  (1.0/(1.0 + b**2))
gammaNNF  =  (b**2/(1.0 + b**2))

alphaNFF  =  alphaNNF
betaNFF   =  betaNNF
gammaNFF  =  gammaNNF

NNF  =    [[cosA*alphaNNF,sinA*alphaNNF,sinA*alphaNNF,-cosA*alphaNNF,0.0,0.0],
           [betaNNF,0.0,0.0,betaNNF,(1.0-betaNNF),0],
           [gammaNNF,0.0,0.0,gammaNNF,(1.0-gammaNNF)*Cx,(1.0-gammaNNF)*Cy]]

#get the dimension
f_dim_steem  =   lambda D : np.abs(alphaNNF)**D +  np.abs(betaNNF)**D +  np.abs(gammaNNF)**D - 1
D_Steem      =   optimize.root(f_dim_steem, 1.8,method='lm').x
```

Make a picture of our triangle to make sure we got it right

```python
IFS_STEEM = se.convertIFS(NNF)

Steem_mom = se.mom_wrap_Steem(IFS_STEEM,D_Steem,n=21,it = 30)

img, _ = se.chaos(IFS_STEEM,[1/3,1/3,1/3], it = 100000, imgxy=500)
img
```

Make a function for our operator $\Phi$, and wrap it so we can form pictures for our minimisation

```python
def Phi_wrap(x,D,it = 1,n=6, sim = False ):    
    assert len(x)%6==0
    
    N = len(x)
    
    a_l = x[:N//6]
    
    b_l = x[1*(N//6):2*(N//6)]
    
    c_l = x[2*(N//6):3*(N//6)]
    
    d_l = x[3*(N//6):4*(N//6)]
    
    e_l = x[4*(N//6):5*(N//6)]
    
    f_l = x[5*(N//6):]
    
    if sim:
    
        ad_l = (np.abs(a_l)+np.abs(d_l))/2

        a_l = np.sign(a_l)*ad_l

        d_l = np.sign(d_l)*ad_l

        bc_l = (np.abs(b_l)+np.abs(c_l))/2

        b_l = np.sign(b_l)*bc_l

        c_l = np.sign(c_l)*bc_l
    
    prob = np.power(np.sqrt(np.abs(a_l*d_l - b_l*c_l)),D-1)
    
    global p_l
    
    p_l =  prob/(np.sum(prob))  
    

    global k

    if k % 20 == 0:
        print('Current maps:')
        img, _ = se.chaos(x,p_l,it = 20000)
        plt.imshow(img)
        plt.show()


    k = k + 1
    
    return mf.Phi(a_l,b_l,c_l,d_l,e_l,f_l,p_l,it = it,n=n)
```

Make a function to minimise and see that our IFS for the Steemson triangle makes the cost `small`. 

Should the reader desire, the `similitude' line can be changed to True. If this is done, at each step in the minimisation, the affine map is projected to its closest similitude function in the induced $L^2$ norm.

```python
k = 0

#can select to project to similitude transformation
similitude = False

def cost_Steem(x):
    
    return Phi_wrap(x,D_Steem,it = 1,n=len(Steem_mom),sim = similitude)@Steem_mom - Steem_mom

print(IFS_STEEM)

cost_Steem(IFS_STEEM)    
```

This starting point corresponds to a Sierpinski triangle with orientations with no flips. The Steemson triangle is also known as a FNN triangle, as it contains two maps which do not use conjugation and one that does.

We now make a `good` starting point. This is still a Sierpinski triangle, but with the correct map oreientations to match a Steemson.

```python
#start Steemson Sierpinski attempt
a  =  1
b  =  1

imgxy = 500

#co ordinates of triangles third vertex
Cy  =  (0.5*np.sqrt(-1 + 2*a**2 - a**4 + 2*b**2 + 2*a**2*b**2 - b**4))
Cx  =  (0.5*(1 - a**2 + b**2))

#angles for roatation
cosA  =  (Cx/b)
sinA  =  (Cy/b)
cosB  =  ((1.0-Cx)/a)
sinB  =  (Cy/a)

#scaling factors 
alphaNFF  =  (b)/((a**2 + b**2))
betaNFF   =  (a)/(a**2 + b**2)
gammaNFF  =  -1.0*(1.0-a**2-b**2)/(a**2 + b**2)

alphaNNF  =  (b/(1.0 + b**2))
betaNNF   =  (1.0/(1.0 + b**2))
gammaNNF  =  (b**2/(1.0 + b**2))

alphaNFF  =  alphaNNF
betaNFF   =  betaNNF
gammaNFF  =  gammaNNF

NNF  =    [[cosA*alphaNNF,sinA*alphaNNF,sinA*alphaNNF,-cosA*alphaNNF,0.0,0.0],
           [betaNNF,0.0,0.0,betaNNF,(1.0-betaNNF),0],
           [gammaNNF,0.0,0.0,gammaNNF,(1.0-gammaNNF)*Cx,(1.0-gammaNNF)*Cy]]

x0 = se.convertIFS(NNF)
```

```python
k=0
t = time.time()
sol_Steem = least_squares( cost_Steem, x0 , bounds=(-1,1) , verbose = 2)
t =  time.time() - t
```

As we can see, we found the global minimum of the function and recovered our Steemson triangle

```python
img,_ = se.chaos(sol_Steem.x,p_l, it = 100000, imgxy=500)
img
```
